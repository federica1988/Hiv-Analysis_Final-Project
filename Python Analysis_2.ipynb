```python
import pandas as pd

# Caricare il file Excel
file_path = r'C:\Users\Security\Desktop\prova\Clean.xlsx'
df = pd.read_excel(file_path)

# Filtrare la colonna Age per la classe 'Age 0-14'
df_filtered_age = df[df['Age'].str.contains('Age 0-14')]

# Escludere la categoria 'Global' dalla colonna Country/Region
df_filtered_country = df_filtered_age[df_filtered_age['Country/Region'] != 'Global']

# Filtrare la colonna Indicator per 'Estimated number of annual AIDS-related deaths'
df_filtered_indicator = df_filtered_country[df_filtered_country['Indicator'] == 'Estimated number of annual AIDS-related deaths']

# Filtrare la colonna Sex per 'Both'
df_filtered_sex = df_filtered_indicator[df_filtered_indicator['Sex'] == 'Both']

# Modifica dei valori non numerici ininfluenti ai fini dell'analisi
df_filtered_sex.loc[df_filtered_sex['Value'] == '<100', 'Value'] = 99
df_filtered_sex.loc[df_filtered_sex['Value'] == '<200', 'Value'] = 199
df_filtered_sex.loc[df_filtered_sex['Value'] == '<500', 'Value'] = 499

# Convertire la colonna 'Value' da stringa a float usando .loc
df_filtered_sex.loc[:, 'Value'] = df_filtered_sex['Value'].astype(float)

# Filtrare gli anni dal 2012 al 2022
df_filtered_years = df_filtered_sex[df_filtered_sex['Year'].between(2012, 2022)]

# Raggruppare per paese e somma i valori per tutto il periodo
df_grouped = df_filtered_years.groupby('Country/Region')['Value'].sum().reset_index()

#'Value' di tipo float
df_grouped['Value'] = df_grouped['Value'].astype(float)

# Top 10 paesi in base alla somma dei valori
top_10_countries = df_grouped.nlargest(10, 'Value')

# Mostrare il risultato
print(top_10_countries)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dati per "Reported number of children receiving ART"
data_reported_art = {
    'Country/Region': ['Eastern and Southern Africa']*11 + ['Sub-Saharan Africa']*11 + ['West and Central Africa']*11,
    'Year': list(range(2012, 2023)) * 3,
    'Value': [459000, 514000, 585000, 629000, 671000, 693000, 662000, 671000, 635000, 611000, 598000,
              524000, 585000, 668000, 723000, 779000, 816000, 785000, 799000, 772000, 753000, 742000,
              65100, 71000, 83400, 93800, 107000, 123000, 123000, 129000, 137000, 142000, 145000]
}

# Dati per "Estimated number of annual new HIV infections"
data_estimated_infections = {
    'Country/Region': ['Eastern and Southern Africa']*11 + ['Sub-Saharan Africa']*11 + ['West and Central Africa']*11,
    'Year': list(range(2012, 2023)) * 3,
    'Value': [140000, 140000, 130000, 110000, 110000, 96000, 82000, 76000, 70000, 65000, 60000,
              220000, 200000, 190000, 170000, 170000, 160000, 140000, 130000, 130000, 120000, 110000,
              73000, 66000, 61000, 60000, 59000, 61000, 62000, 58000, 57000, 53000, 50000]
}

# Convertire i dati in DataFrame
df_reported_art = pd.DataFrame(data_reported_art)
df_estimated_infections = pd.DataFrame(data_estimated_infections)

# Funzione per creare grafico a linea per un paese specifico
def create_line_chart(country):
    plt.figure(figsize=(10, 6))
    
    # Filtrare i dati per il paese specifico
    df_art_country = df_reported_art[df_reported_art['Country/Region'] == country]
    df_infections_country = df_estimated_infections[df_estimated_infections['Country/Region'] == country]
    
    # Tracciare il grafico per "Reported number of children receiving ART"
    plt.plot(df_art_country['Year'], df_art_country['Value'], marker='o', label='Reported number of children receiving ART')
    
    # Aggiungere le annotazioni per i punti
    for x, y in zip(df_art_country['Year'], df_art_country['Value']):
        plt.annotate(f'{y}', xy=(x, y), textcoords='offset points', xytext=(0,5), ha='center')
    
    # Tracciare il grafico per "Estimated number of annual new HIV infections"
    plt.plot(df_infections_country['Year'], df_infections_country['Value'], marker='o', label='Estimated number of annual new HIV infections')
    
    # Aggiungere le annotazioni per i punti
    for x, y in zip(df_infections_country['Year'], df_infections_country['Value']):
        plt.annotate(f'{y}', xy=(x, y), textcoords='offset points', xytext=(0,5), ha='center')
    
    # Impostare il titolo e le etichette
    plt.title(f'Trend of HIV Infections and ART in {country}')
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.xticks(range(2012, 2023))
    
    # Mostrare il grafico
    plt.show()

# Creare i grafici per ciascun paese
create_line_chart('Eastern and Southern Africa')
create_line_chart('Sub-Saharan Africa')
create_line_chart('West and Central Africa')
```
